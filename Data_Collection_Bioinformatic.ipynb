```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# We use EGFR data for target protein reading
target_EGFR = new_client.target
target_query_EGFR = target_EGFR.search('EGFR')
targets_EGFR = pd.DataFrame.from_dict(target_query_EGFR)
targets_EGFR
```

```python
# From the list of targets we picked CHEMBL203
selected_target_EGFR = targets_EGFR.target_chembl_id[3]
selected_target_EGFR
```

```python
# CheMBL203 Bioactivity data retrieval
# we picked every data that are reported as pCheMBL value
# CheMBL: allows a number of roughly comparable measures of half-maximal response concentration/potency/affinity to be compared on a negative logarithmic scale
# ex: -Log(molar IC50, XC50, EC50, AC50, Ki, Kd or Potency)
activity_EGFR = new_client.activity
res_EGFR = activity_EGFR.filter(target_chembl_id=selected_target_EGFR).filter(standard_type="IC50")
```

```python
# Data reading
df_EGFR = pd.DataFrame.from_dict(res_EGFR)
df_EGFR
```

```python
# Storing the raw data for observation
df_EGFR.to_csv('1_EGFR_Bioactivity_Raw_Data.csv', index=False)
filename_EGFR = '1_EGFR_Bioactivity_Raw_Data.csv'
```

```python
# reading the preprocessed data for labeling
# standard_value >= 10000 -> active
# standard_value <= 10000 -> inactive
df4_EGFR = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/UPDATED CODE & DATA Bio Colab (Final)/data/2_EGFR_Bioactivity_Preprocessed_Data.csv')
df4_chy = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/UPDATED CODE & DATA Bio Colab (Final)/data/2_Chymotrypsin_Bioactivity_Preprocessed_Data.csv')
```

```python
# EGFR
bioactivity_threshold_EGFR = []
for i in df4_EGFR.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold_EGFR.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold_EGFR.append("active")
  else:
    bioactivity_threshold_EGFR.append("intermediate")
```

```python
# Chy
bioactivity_threshold_chy = []
for i in df4_chy.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold_chy.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold_chy.append("active")
  else:
    bioactivity_threshold_chy.append("intermediate")
```

```python
bioactivity_class_EGFR = pd.Series(bioactivity_threshold_EGFR, name='class')
df5_EGFR = pd.concat([df4_EGFR, bioactivity_class_EGFR], axis=1)
df5_EGFR
```

```python
bioactivity_class_chy = pd.Series(bioactivity_threshold_chy, name='class')
df5_chy = pd.concat([df4_chy, bioactivity_class_chy], axis=1)
df5_chy
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
# Lipinski descriptors calculation
# Molecular weight < 500 Dalton
# Octanol-water partition coefficient (LogP) < 5
# Hydrogen bond donors < 5
# Hydrogen bond acceptors < 10

def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
df_lipinski_EGFR = lipinski(df_clean_smiles_EGFR.canonical_smiles)
df_lipinski_EGFR
```

```python
df_lipinski_chy = lipinski(df_clean_smiles_chy.canonical_smiles)
df_lipinski_chy
```

```python
# combine 2 dataframes which are the curated data and the result of lipinski descriptor calculation
df_combined_EGFR = pd.concat([df_EGFR,df_lipinski_EGFR], axis=1)
df_combined_EGFR
```

```python
# combine 2 dataframes which are the curated data and the result of lipinski descriptor calculation
df_combined_chy = pd.concat([df_chy,df_lipinski_chy], axis=1)
df_combined_chy
```

```python
# convert IC50 to pIC50 by converting the IC50 value to negative logarithmic value

import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop(columns=['standard_value_norm'])

    return x
```

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop(columns=['standard_value'])

    return x
```

```python
df_norm_EGFR = norm_value(df_combined_EGFR)
df_norm_EGFR
```

```python
df_norm_chy = norm_value(df_combined_chy)
df_norm_chy
```
